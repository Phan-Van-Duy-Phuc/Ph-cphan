```
import numpy as np
A = np.array([[1,3,5],[5,4,2],[4,6,8]])
B = np.array([[3,5,2],[7,1,-6],[8,6,-4]])
print("ma tran A la",A)
print("ma tran B la",B)
C = A+B
print("tong hai ma tran =",  C)
D = A.dot(B)
print("tich hai ma tran =",  D)
A = np.array([[1,3,5],[5,4,2],[4,6,8]])
print("chuyen vi cua ma tran A la",A.transpose())
```

```
  #- Đồ thị hình column:
import matplotlib.pyplot as plt 
import numpy as np  
divisions = ["Dừa", "Táo", "Cam", "Xoài","Vải"]
divisions_average_marks = [80,63,72,67,90]
plt.bar(divisions, divisions_average_marks, color = 'blue')
plt.title(" Số lượng hoa quả trung bình mỗi người ăn trong 1 năm")
plt.xlabel("Các loại hoa quả")
plt.ylabel("%")
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
divisions = ["SS S20+", "SS Z Flip", "SS Z Fold 2", "SS S10+", "SS A71"]
division_average_marks = [70,50,50,70,40]
boys_average_marks = [60,30,80,38,56]

index = np.arange(5)
width = 0.20

plt.bar(index, division_average_marks, width, color= 'green', label= 'Việt Nam')
plt.bar(index+ width, boys_average_marks, width, color = 'red', label= 'Mỹ')
plt.ylabel("%")
plt.xlabel("Các loại điện thoại samsung")
plt.xticks(index+ width/2, divisions)        
plt.legend(loc = 'best') 
plt.show()
```
